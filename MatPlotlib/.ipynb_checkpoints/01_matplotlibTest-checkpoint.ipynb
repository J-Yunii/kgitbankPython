```python
# evenly sampled time at 200ms intervals
# 0.2씩 증가 0-5범위에서
t = np.arange(0., 5., 0.2)
t
```

```python
# x값, y값, 선모양
# red dashes, blue square, green triangle
plt.plot(t,t,'r--',t, t**2,'bs',t, t**3,'g^')
```

```python
# arange(50) =>데이터 50개
# randn: 정규분포형태 데이터
data = {'a': np.arange(50),
       'c':np.arange(50),
       'd':np.random.randn(50)}

# 임의로 데이터 값을 형성
# abs는 절댓값
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
data
```

```python
# 산포도, 큰 값은 크게 작은 값은 작게
# c: 컬러, marker: 마커기호
plt.scatter('a','b',c='c',s='d',marker='^',data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.minorticks_on() # 보조눈금선
plt.show
```

```python
names = ['group_a','group_b','group_c']
values = [1,10,100]

# 너비, 높이
plt.figure(figsize=(9,3))

# 큰 테두리 안에 서브 그래프 3개가 들어간다.
# 이때 131이라는 뜻은 1행 3칸짜리에서 1번째 위치
plt.subplot(234)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```
