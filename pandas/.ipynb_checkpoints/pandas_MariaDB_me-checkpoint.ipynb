```python
import pymysql
connection=pymysql.connect(host='maria',
                            user='root',
                            password='qwer1234',
                            db='test',
                            charset='utf8mb4',
                            cursorclass=pymysql.cursors.DictCursor)

try:
    with connection.cursor() as cursor:
        sql='''
           drop table if exists users;
        '''
        cursor.execute(sql)
        connection.commit()

    with connection.cursor() as cursor:
        sql='''
           create table if not exists users(
            userid varchar(20) primary key,
            userpw varchar(20) not null,
            username varchar(20) not null,
            userage int,
            usermail varchar(50),
            useradd varchar(50),
            usergender varchar(20),
            usertel varchar(20));
        '''
        cursor.execute(sql)
        connection.commit()

    with connection.cursor() as cursor:
        sql='''
            insert into users values('gildong1','1234','길동',33,
            'gildong@gmail.com','부산시 남구','male','010-1234-1234');
        '''
        cursor.execute(sql)
        connection.commit()
    
    with connection.cursor() as cursor:
        sql='''
            insert into users values('dong1','1234','정윤',23,
            'jungdong@gmail.com','부산시 남구','male','010-1234-1234');
        '''
        cursor.execute(sql)
        connection.commit()

    with connection.cursor() as cursor:
        sql='''
            select  * from users;
        '''
        cursor.execute(sql)
        result=cursor.fetchall()
        print(result)
finally:
    connection.close()
```

# 3. read_sql()
sql 쿼리 또는 db 테이블을 DataFrame으로 읽습니다.
sql 쿼리는 read_sql_query로 라우팅되고 db 테이블 이름은 read_sql_table로 라우팅됩니다.

```python
import pymysql
import pandas as pd
import sqlalchemy
db =pymysql.connect(host='maria',
                            user='root',
                            password='qwer1234',
                            db='test',
                            charset='utf8mb4',
                            cursorclass=pymysql.cursors.DictCursor)

sql = 'select * from users'
# SQL query to be executed or a table name.
df = pd.read_sql(sql,db)
print(type(df))
df
```

```python
from sqlalchemy import create_engine
# 'mysql+pymysql://root' + '비밀번호' + 'mariadb명칭/db이름'
engine = create_engine('mysql+pymysql://root'+':qwer1234'+'@maria/test',encoding='utf-8')

data = engine.execute('SELECT * FROM users').fetchall()
print(type(data))
print(data)
```

```python
sql = "select * from users"
df = pd.read_sql(sql,engine)
print(type(df))
df
```

```python
df = pd.read_sql('users',engine) # read_sql_table()로 라우팅
print(type(df))
df
```

```python
# 테이블 이름만 받음
data = pd.read_sql_table("users", engine)
data
```

```python
df = pd.read_sql_query('users',engine) # 오류 쿼리만 입력가능
df
```

```python
pd.read_sql_query('select * from users where userage < 10;', engine)
```

```python
# 데이터 프레임을 json으로 변환, 한글은 유니코드 형태로 나옴
df.to_json()
```

```python
import sqlalchemy as sa
metadata = sa.MetaData()
data_table = sa.Table('users', metadata,
                     sa.Column('userid', sa.String, primary_key = True),
                     sa.Column('userpw',sa.String),
                     sa.Column('username',sa.String),
                     sa.Column('userage',sa.Integer),
                     sa.Column('usermail',sa.String),
                     sa.Column('useradd',sa.String),
                     sa.Column('usergender',sa.String),
                     sa.Column('usertel',sa.String))

# 컬럼.userage가 10보다 큰 것을 선택
pd.read_sql_query(sa.select([data_table]).where(data_table.c.userage>10).order_by(data_table.c.userid),engine)
```

```python
import urllib.request
import urllib.parse

# 긁어올 url 선택
# https://search.naver.com/search.naver?sm=top_hty&fbm=1&ie=utf8&query=%EC%B4%88%EC%BD%9C%EB%A6%BF
api = 'https://search.naver.com/search.naver'
# url 뒤에 붙는 값을 => 딕셔너리 형태로
# 딕셔너리 값을 url로 전환해줌
values = {
    'sm':'top_hty',
    'fbm':'1',
    'ie':'utf8',
    'query':'초콜릿'
}
query = input('검색어 입력')
values['query'] = query


# parse: 딕셔너리 값을 url로 전환해줌
params=urllib.parse.urlencode(values)
print(params)
url=api+'?'+params
data=urllib.request.urlopen(url).read()
# print(data)
text=data.decode('utf-8')
print(text)
```
