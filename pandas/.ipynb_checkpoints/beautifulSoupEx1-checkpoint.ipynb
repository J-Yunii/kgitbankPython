```python
# 환율 크롤링하기
import urllib.request
from bs4 import BeautifulSoup 

# url에 한글있으면 parse써서 바꿔주기
url = 'https://search.naver.com/search.naver?sm=top_hty&fbm=0&ie=utf8&query=%ED%99%98%EC%9C%A8'


html_doc = urllib.request.urlopen(url).read()
#print(html.decode('utf-8'))
soup = BeautifulSoup(html_doc,'html.parser')
print(soup.prettify())
```

```python
table=soup.find_all('table','rate_table_info')
```

```python
# dollardw=soup.select('tbody>.dw>th>a>span')
# dollarup=soup.select('tbody>.up>th>a>span')
# print(dollardw, dollarup)
```

```python
results = soup.select('th>a>span')
results
```

```python
for result in results:
    # 객체로 나오므로 list로 변환해서 출력
    print(list(result.strings))
    # contents로 가져와도 괜찮음
    print(result.contents[0])
```

```python
dollars = soup.select('tr>td>span')
dollars
```

```python
# 방법 1
i=0
for dollar in dollars:
    if(i%4==0):
        print(dollar.string)
    i=i+1
```

```python
# 방법 2
results=soup.select('tr')
for result in results:
    print(result.contents[3:4])
```

```python
# 방법 3
results=soup.select('td>span')
# print(results)
for result in results:
    #print(type(result))
    #print(result.has_attr('class'))
    if not result.has_attr('class'):
        # has_attr: 속성값을 가지고 있는지 체크해 줌
        print(result.string)
```
